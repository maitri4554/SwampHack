```python
#idea behind kmeans: https://medium.com/analytics-vidhya/k-means-clustering-with-python-beginner-tutorial-45a44c34e7f 
#csv file: https://figshare.com/articles/dataset/Main_Dataset_for_Evolution_of_Popular_Music_USA_1960_2010_/1309953 
import pandas as pd
import numpy as np
import random
from sklearn.cluster import KMeans
```

```python
a = pd.read_csv("EvolutionPopUSA_MainData.csv")
X = a.drop(labels=["artist_name","decade", "fiveyear", "track_name","quarter","first_entry", "era", "cluster"], axis=1)
Y = a['artist_name_clean']
```

```python
X = X.drop(X.columns[1:32], axis=1)
X
```

```python
a = a.drop(labels=["cluster"], axis=1)
a
```

```python
kmeans = KMeans(n_clusters = 700)
kmeans.fit(X)
clusters = kmeans.predict(X)
a["cluster"]=clusters
```
